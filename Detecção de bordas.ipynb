```python
import numpy as np
import scipy
import cv2
import time
from skimage import feature
from skimage.filters import sobel,sobel_h, sobel_v,apply_hysteresis_threshold
from skimage.io import imread,imsave
from skimage.transform import hough_circle,hough_circle_peaks
from skimage.draw import circle_perimeter,ellipse_perimeter
from matplotlib import pyplot as plt
```

```python
olho = imread("D:/Ubiris_GRAY/Sessao_1/23/Img_23_1_1.jpg",as_gray=True)
#olho = imread("D:/MMU/iris5b.bmp",as_gray=True)
ori = olho.copy()
#olho = cv2.medianBlur(olho,7)
olho = cv2.blur(olho,(5,5))
#olho = cv2.GaussianBlur(olho,(5,5),0)
borda1 = feature.canny(olho)
borda2 = feature.canny(olho,2)
```

```python
fig,axi = plt.subplots(nrows=2,ncols=2,figsize=(15,11),sharex=True,sharey=True)
(ax1,ax2,ax3,ax4)=np.extract(axi.ravel(),axi)

ax1.axis('off')
ax1.set_title('Imagem original')
ax1.imshow(ori,cmap='gray')

ax2.axis('off')
ax2.set_title('Imagem suavizada')
ax2.imshow(olho,cmap='gray')

ax3.axis('off')
ax3.set_title('Sigma 0')
ax3.imshow(borda1,cmap='gray')

ax4.axis('off')
ax4.set_title('Sigma 2')
ax4.imshow(borda2,cmap='gray')
```
